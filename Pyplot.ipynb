`Pyplot` proporciona la funcionalidad básica para generar gráficas aceptables para una publicación científica o técnica.

Hay que importar los paquetes `matplotlib.pyplot` y `numpy`, y conviene declarar alias (`plt` y `np` están muy extendidos).

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

Generamos una gráfica a partir de una serie de valores para el eje X y para el eje Y.

```python
xs1=np.arange(0,2*np.pi,0.02)  # Generacion de un array
ys1=np.sin(xs1)  # Actuacion sobre el array completo
plt.plot(xs1,ys1)  # Generacion de la grafica
plt.savefig('figpyplot3.pdf')
plt.show()  # La grafica se muestra
```

Para guardar la gráfica en un fichero se usa la orden `plt.savefig('prueba.pdf')`, donde 'prueba' es el nombre del fochero y la extensión '.pdf' indica el formato de salida. Este formato puede ser también '.png' y '.eps' entre otros. 

Esta instrucción hay que ponerla antes de visualizar la gráfica mediante `plt.show()`. De lo contrario genera una gráfica vacía.

El dibujo de puntos se hace con la función `scatter`. Tiene sintaxis similar a `plot`. Los datos experimentales deben representarse con puntos.

```python
xsp=np.arange(10)
ysp=xsp**2
plt.scatter(xsp,ysp)
plt.show()
```

## Gráficas científicas

* Etiquetas: `label`, `legend`, con `loc=` para la posición: 1 esquina superior derecha, 2, ... 4 esquina inferior derecha en sentido antihorario
* Título: `title`
* Leyenda en los ejes X e Y: `xlabel`, `ylabel`, con `fontsize`
* Cuadriculado: `grid`
* Límites en cada eje: `xlim`, `ylim` (despues de `plot` y antes de `show`)

```python
xs1=np.arange(0,2*np.pi,0.02)
ys1=np.sin(xs1)
ys3=np.cos(xs1)

plt.plot(xs1,ys1,label='Sin(x)')
plt.plot(xs1,ys3,label='Cos(x)')
plt.legend(loc=1)
plt.xlabel('Radianes')
plt.ylabel('Amplitud')
plt.xlim(-1,7)
plt.ylim(-2,2)
plt.grid()
plt.savefig('figpyplot4.pdf')
plt.show()
```

### Uso de comandos de latex para escribir ecuaciones en las gráficas

Necesita tener instalado un paquete de Latex. Hace uso de la notación matemática, anteponiendo la letra `r` antes de la cadena, y escribiendo la expresión matemática en Latex.

```python
plt.rc('text',usetex=True) 
x=np.linspace(-10,10,1001)
for i in range(1,5):
    y=x**i*np.sin(x)
    y/=max(y)  # se normaliza al valor maximo
    plt.plot(x,y,label=r'$x^{} \sin(x)$'.format(i))
plt.legend(loc=8)  # La leyenda inferior centrada
plt.savefig('figpyplot5.pdf')
plt.show()
```

### Marcas en las líneas

Para incluir algún símbolo asociado a cada tanda de datos se puede usar la opción `marker`. Es interesante cuando se representan medidas experimentales. Los símbolos asociados son:

* `'.'` punto
* `'o'` círculo 
* `'+'` más 
* `'x'` x 
* `'d'` diamantes
* `'v'` triángulo invertido
* `'^'` triángulo hacia arriba
* `'s'` cuadrado 
* `'*'` asterisco

```python
x=np.linspace(0,1,11)
y=x
plt.plot(x,y,marker='D')
plt.savefig('figpyplot6.pdf')
plt.show()
```

## Color, estilo y ancho de las líneas

Hay diferentes maneras de especificar el color, estilo y ancho de las líneas. La manera más sencilla es a través de los parámetros `c`, `ls`, `lw`, que son abreviaturas de `color`, `linestyle`, `linewith`, que también se pueden usar como parámetros con igual funcionalidad. Las especificaciones son:

* `'b'` azul
* `'g'` verde 
* `'r'` rojo 
* `'c'` cyan
* `'m'` magenta
* `'y'` amarillo 
* `'k'` black 
* `'w'` blanco

* `'-'` sólida 
* `'--'` dashed 
* `':'` dotted 
* `'-.'` dash-dot 

Para el ancho se define un real, donde 1 corresponde a un valor normal.

```python
x=np.linspace(0,1,11)
y=x
plt.plot(x,y,c='r',ls='--',lw=0.5)
plt.savefig('figpyplot7.pdf')
plt.show()
```

## Cómo crear una figura con dos ejes

```python
import numpy as np
import matplotlib.pyplot as plt
zps=np.linspace(0,3,200)  
plt.grid()
plt.xlabel("z'")
plt.ylabel('z^2')  
l1=plt.plot(zps,zps**2,c='b',ls=':') # Primer eje
plt.twinx()
l2=plt.plot(zps,zps/2.,c='r',ls='--') # Segundo eje
plt.ylabel('z/2')
l12=l1+l2
plt.legend(l12,['z^2','z/2'])
plt.savefig('figpyplot8.pdf')
plt.show()
```
